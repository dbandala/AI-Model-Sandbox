```python
import json
import os
from pathlib import Path
import pandas as pd
```

```python
from dotenv import load_dotenv
from langchain_core.vectorstores import InMemoryVectorStore
from langchain_openai import OpenAIEmbeddings
```

```python
load_dotenv()
```

```python
# Create an OpenAI embeddings instance
embeddings_model = OpenAIEmbeddings(
    model="text-embedding-3-large", api_key=os.getenv("OPENAI_API_KEY")
)
# Create an vector store instance
vector_store = InMemoryVectorStore(embeddings_model)
```

```python
embeddings_model.embed_query("Hello, world!")
```

```python
from langchain_core.documents import Document

document_1 = Document(id="1", page_content="foo", metadata={"baz": "bar"})
document_2 = Document(id="2", page_content="thud", metadata={"bar": "baz"})
document_3 = Document(id="3", page_content="i will be deleted :(")

documents = [document_1, document_2, document_3]
vector_store.add_documents(documents=documents)
```

```python
similarity_search_results = vector_store.similarity_search_with_score("faa", 5)
```
